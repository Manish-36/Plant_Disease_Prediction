## Importing Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Data Preprocessing

### Training Image Preprocessing

```python
training_set = tf.keras.utils.image_dataset_from_directory(
    'train',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

### Validation Image Preprocessing

```python
validation_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
training_set = training_set.prefetch(tf.data.AUTOTUNE)
validation_set = validation_set.prefetch(tf.data.AUTOTUNE)
```

### To avoid overshooting

1. Choose small learning rate .... instead of 0.001 choose 0.0001
2. There maybe chance of underfitting so increase the number of neurons
3. Add more convolutional layer to extract features that may not have captured by the model

## Building Model

```python
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
```

## Building Convolutional Layer

```python
model.add(Conv2D(filters=32,kernel_size=3,padding='same',activation='relu',input_shape=[128,128,3]))   
model.add(Conv2D(filters=32,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2)) #poolsize is size of maxpool filter 2x2 and strides are the number of pixels by which the pool will move
#We will get 32 different feature maps when we apply 32 filters
#kernel size is the filter size which 3x3 here
#input shape is the size of the image declared earlier along with rgb [128x128,3]
```

```python
model.add(Conv2D(filters=64,kernel_size=3,padding='same',activation='relu',input_shape=[128,128,3]))   
model.add(Conv2D(filters=64,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2)) 
```

```python
model.add(Conv2D(filters=128,kernel_size=3,padding='same',activation='relu',input_shape=[128,128,3]))   
model.add(Conv2D(filters=128,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2)) 
```

```python
model.add(Conv2D(filters=256,kernel_size=3,padding='same',activation='relu',input_shape=[128,128,3]))   
model.add(Conv2D(filters=256,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2)) 
```

```python
model.add(Conv2D(filters=512,kernel_size=3,padding='same',activation='relu',input_shape=[128,128,3]))   
model.add(Conv2D(filters=512,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2)) 
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Flatten())  #it is addition of the flattening layer at the end of convolution layers
```

```python
model.add(Dense(units=1500,activation='relu')) #it is the dense layer/hidden layer added to the model(cnn)  units-->no. of neurons  
```

```python
model.add(Dropout(0.4))
```

```python
# Output layer......units are 38 as there are 38 classes if you see the output of training_set
model.add(Dense(units=38,activation='softmax'))
```

### Compiling the Model

```python
model.compile(optimizer=tf.keras.optimizers.legacy.Adam(learning_rate=0.0001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()  #to see all the parameters of the model
```

### Training our Model

```python
training_history = model.fit(x=training_set,validation_data=validation_set,epochs=10)
```

# Model Evaluation

```python
# Model evaluation on training set
train_loss,train_acc = model.evaluate(training_set)
```

```python
print(train_loss,train_acc)
```

```python
# Model evaluation on validation set
val_loss,val_acc = model.evaluate(validation_set)
```

```python
print(val_loss,val_acc)
```

# Saving our model

```python
model.save("trained_model.keras")
```

### Some other metrics for model evaluation

```python
# Temporary dataset just to get class names
temp_validation = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

class_names = temp_validation.class_names
print(class_names)
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=False,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
y_pred = model.predict(test_set)
y_pred
```

```python
y_pred.shape
```

```python
# y_pred gives 38 probabilities for each images
# we have to find out the highest probability amongst the 38 probabilities to succesfully predict
predicted_categories = tf.argmax(y_pred,axis=1)
#tf.argmax returns the index of maximum value from y_pred  (axis=1 means it performs in column wise fashion...axis=0 means row wise)
```

```python
predicted_categories
```

```python
true_categories = tf.concat([y for x,y in test_set],axis=0)  #all the ys of test set are concatenated in row wise fashion
true_categories
```

```python
Y_true = tf.argmax(true_categories,axis=1)
Y_true
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_true,predicted_categories,target_names=class_names))
```

```python
cm = confusion_matrix(Y_true,predicted_categories)
cm,cm.shape
```

### Confusion Matrix Visualization

```python
plt.figure(figsize=(40,40))
sns.heatmap(cm,annot=True,annot_kws={'size':10})
plt.xlabel("Predicted Class",fontsize=20)
plt.ylabel("Actual Class",fontsize=20)
plt.title("Plant Disease Prediction Confusion Matrix",fontsize=25)
plt.show()
```
